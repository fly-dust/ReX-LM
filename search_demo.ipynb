```python
from transformers import Qwen2Tokenizer
import bm25s
from bm25s.hf import BM25HF

def retrieve_nearest_datapoints(query, k=3, rex_index_name="Llama-3-Magpie-Pro-1M-v0.1"):
    hf_repo_name = f"yuchenlin/BM25S_index_{rex_index_name}"
    retriever = BM25HF.load_from_hub(
        hf_repo_name, revision="main", load_corpus=True
    )
    
    query_tokens = bm25s.tokenize(query, show_progress=False)
    results, scores = retriever.retrieve(query_tokens, k=k, show_progress=False)
    
    retrieved_data = []
    for i in range(results.shape[1]):
        doc, score = results[0, i], scores[0, i]
        rex_query = doc["query"]
        rex_response = doc["response"]
        retrieved_data.append({
            "query": rex_query,
            "response": rex_response,
            "score": score
        })
    
    return retrieved_data

def main():
    queries = [
        "What are some common parameters or metrics used in measuring the performance of a machine learning model?"
    ]

    for query in queries:
        print(f"\nProcessing query: {query}")
        results = retrieve_nearest_datapoints(query)
        
        for i, result in enumerate(results, 1):
            print(f"Result {i}:")
            print(f"Rex Query: {result['query']}")
            print(f"Rex Response: {result['response']}")
            print(f"Score: {result['score']}")
            print("---")

if __name__ == "__main__":
    main()
```
